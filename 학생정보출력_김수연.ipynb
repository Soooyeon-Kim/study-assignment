```python
while True:
        number = input("학번을 입력해주세요.")
        #학번
        snumber = number

        #학년 
        #21년도 이후 학번을 입력했을 경우 오류
        if int(number[0:2]) < 19 :
            grade = 4
        elif int(number[0:2]) < 20 :
            grade = 3
        elif int(number[0:2]) < 21 :
            grade = 2
        elif int(number[0:2]) < 22 :
            grade = 1
        else:
            print("학번을 잘못 입력하셨습니다.")
            continue
            
        
        #성별 
        # 1과 2 이외의 숫자를 입력했을 경우 오류
        if number[2] == "1" :
            gender = "여성"
        elif number[2] == "2" :
            gender = "남성"
        else:
            print("학번을 잘못 입력하셨습니다.")
            continue
            


        #학과
        #범위 이상의 숫자를 입력했을 경우 오류
        if int(number[5:7]) < 30 :
            dp = "컴퓨터학과"
        elif int(number[5:7]) < 90 :
            dp = "정보통계학과"
        elif int(number[5:7]) < 100 :
            dp = "문헌정보학과"
        else:
            print("학번을 잘못 입력하셨습니다.")
            continue

        #휴학여부
        if len(number)== 7 and number[-1] != 's':
            h = '해당없음'
        elif len(number)== 8 and number[-1] == 's':
            h = '1년 휴학/ 추가 휴학 가능'
        elif len(number)== 9 and number[-2] =='s':
            h = '2년 휴학/ 추가 휴학 불가'    
        else:
            print("학번을 잘못 입력하셨습니다.")
            continue
            

        #나이
        # 1951 ~ 2002년생까지
        age = number[3:5]

        if int(age) > 50 : 
            ageage = 2022-(1900+int(age))
        elif int(age) < 3 :
            ageage = int(age)+22
        else:
            print("학번을 잘못 입력하셨습니다.")
            continue


        print("------------------------------------------------")
        print("학번 : {} ".format(snumber)),
        print("학년 : {} ".format(grade)),
        print("나이 : {}".format(ageage)),
        print("성별 : {}".format(gender)),
        print("학과 : {}".format(dp)),
        print("휴학여부 : {} ".format(h)),
        print("------------------------------------------------")
        
    
```
